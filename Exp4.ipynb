```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_rows',None)
```

```python
df = pd.read_csv('disaster_train.csv');
df.head(10)
```

```python
df.info()
```

```python
all_hashtags =[] 
def extract_hashtag(sentence):
    words = sentence.split()
    for w in words :
        if w.startswith('#'):
            w = w.lower()
            all_hashtags.append(w)

df['text'].apply(extract_hashtag);
print(all_hashtags)
```

```python
hashtag_counts = { };
for tag in all_hashtags:
    if tag in hashtag_counts :
        hashtag_counts[tag]+=1
    else:
        hashtag_counts[tag]=1;
print(hashtag_counts)
```

```python
hashtag_counts = pd.Series(hashtag_counts).sort_values(ascending=False)
hashtag_counts
```

```python
hashtag_counts.head(10).plot(kind = 'bar' , color='steelblue')
plt.title('Top Hashtags')
plt.xlabel('Hashtag')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
